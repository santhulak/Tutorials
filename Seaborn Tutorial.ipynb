# Bar Plot

```python
# load dataset
titanic = sns.load_dataset('titanic')
sns.set_context('paper')
# create plot
sns.barplot(x = 'embark_town', y = 'age', data = titanic,  palette = 'PuRd',
           ci=None
            )
plt.show()
```

```python
sns.barplot(x = 'sex', y = 'survived', hue = 'class', data = titanic,palette = 'hls',
            order = ['male', 'female'],
            capsize = 0.15,
            saturation = 8,
            errcolor = 'gray', errwidth = 2,      
            ci = 'sd')
plt.legend()
plt.show()
```

#     Count plot
          

## histogram across a categorical variable

```python
sns.countplot(x = 'class', hue = 'who', data = titanic, palette ='magma')
plt.title('Survivors')
plt.show()
```

# Point Plot

```python
data = sns.load_dataset("tips") 
sns.pointplot(x="day", y="tip", data=data)
plt.show()
```

# Joint Plot

```python
tips=sns.load_dataset('tips')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg') 
plt.show()
```

# Regplot

## visualize the linear relationship as determined through regression

```python
ax = sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
#ci=None to remove the highlight in the line
ax = sns.regplot(x="total_bill", y="tip", data=tips,ci=None)
```

# Box Plot
## depicting groups of numerical data through the quartiles

- Minimum -bottom black horizontal line of the box plot
- First Quartile or 25% -first black horizontal line of the rectangle shape of the box plot 
- Median (Second Quartile) or 50% - second black horizontal line of the rectangle shape of the box plot
- Third Quartile or 75% - third black horizontal line of rectangle shape
- Maximum - top black horizontal line of the rectangle shape
- outlier data - small diamond shape of the box plot

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

# Heatmap
##  two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colours.

```python
flights=sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
print(flights)
```

```python
sns.heatmap(flights,linewidths=.5,cmap="YlGnBu")
```

```python
flights=sns.load_dataset("flights")
corr=flights.corr()
print(corr)
sns.heatmap(corr,annot=True,linewidths=.5,cmap="YlGnBu")
```

# Cluster map
## plots a matrix dataset as a hierarchically-clustered heatmap

```python
flights=sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
sns.clustermap(flights,linewidths=.5,cmap="coolwarm")
```

# Facetgrid
## matrix of panels defined by row and column by dividing the variables.

```python
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time")
g.map(sns.scatterplot, "total_bill", "tip")
```

# Pair Plot
## creates a grid of Axes of numeric variable in data 

```python
df = sns.load_dataset('iris')
sns.set_style("ticks")
sns.pairplot(df,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```
